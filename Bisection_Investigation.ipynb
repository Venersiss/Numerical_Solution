```python
import math

print("This is my prototype of Bisection Investigation")
equation="x^5 - sin(x) - 0.5"


Xl = 0
Xu = 2
Xm = 0
answer = 0
iteration= 1
tolerance = 0.01
temp=0

def Y(data):      
    return data**5 - math.sin(data) - 0.5


def X(data):
     return data**5 - math.sin(data) - 0.5


def solution(x,y):
    return X(x)*Y(y)

    
def root_estimate(x, y):
    
    return (x+y)/2
    

print("The equation Given is: "+str(equation)+"\n\n")
print("My Xl is: " + str(Xl))
print("My Xl is: " + str(Xu))
print("My f(Xl) is: " + str(X(Xl)))
print("My f(Xu) is: " + str(Y(Xu)))
print("The product of Xl and Xu is: " + str(solution(Xl, Xu)) + "\n"+
      "which is less than 0 which means there is at least"+
      "\n"+ "one root on "+str(Xl)+" and "+str(Xu))
answer = solution(Xl, Xu)

if answer <0:
    for x in range(20):
        print("\n")
        print("Iteration " + str(x + 1))
        Xm = root_estimate(Xl, Xu)
        answer = solution(Xl, Xm)
        temp = X(Xm)
        print("My Xl is: " + str(Xl))
        print("My Xu is: " + str(Xu))
        print("My Xm is: " + str(Xm))
        print("My f(Xm) is: " + str(temp))
        if abs(temp)<tolerance:
            break
        if answer < 0:
            print("Product of Xl and Xm is less than 0 " + str(answer))
            Xl = Xl
            Xu = Xm
        elif answer > 0:
            print("Product of Xl and Xm is more than 0 " + str(answer))
            Xl = Xm
            Xu = Xu
        elif answer == 0:
            print("The root is:" + str(Xm))
            break
    print("\n")
    print("The root is: "+str(Xm))

       
            
        
            
elif answer > 0:
    print("hehe")
    
    
# Initialize a flag to track if the condition has been met



```
