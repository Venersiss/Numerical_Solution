```python
import math
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (0.312 * np.exp(0.895 * x)) + (2*x) + 4

def numerical_derivative(x):
    h = 1e-5  # Small step size (infinitesimal)
    return (f(x + h) - f(x)) / h

def derive(x):
    derivative_value = numerical_derivative(x)
    return derivative_value

def eq(data1, data2): 
    return data1 / data2

def Ea(data1, data2):
    return abs((data1 - data2) / data1) * 100

print("This is Newton-Rhapson Method")
equation = "0.312*e**(0.895 * x) + 2*x + 4"

X1 = 10
tolerance = 0.1
iteration = 0

while True:
    iteration=iteration+1
    print("Iteration: "+str(iteration)+"\n")   
    X2 = X1 - eq(f(X1), derive(X1))
    print("Xi+1 = "+str(X2)+"\n")  
    
    if Ea(X2, X1) <= tolerance:
        
        print("The Absolute Relative Approximate Error is: "+str(Ea(X2, X1)))
        print("it is less than or equal to our tolerance which is: "+str(tolerance))
        print("so we stop \n \n")
        break
        
        
    print("The Absolute Relative Approximate Error is: "+str(Ea(X2, X1)))
    print("it is greater than our tolerance which is: "+str(tolerance))
    print(str(X1)+" "+str(X2))
    print("so we loop again \n \n")
    X1 = X2

print("The root is", X2)

# Plot
x_values = np.linspace(-10, 10, 1000)
y_values = f(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r"$f(x) = "+equation+"$")
plt.scatter(X2, f(X2), color='red', label='Root')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x) ="+equation)
plt.grid(True)
plt.legend()
plt.show()

   


    
    




```
